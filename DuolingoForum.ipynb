````python
import urllib.request
from bs4 import BeautifulSoup as bs
import re
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

import time

# --------------------------------------------------------
# DO NOT CHANGE
#
def get_links(driver, url, wait):
    links=[]
    driver.get(url)
    time.sleep(wait)
    elements = driver.find_elements_by_class_name("_3ZcIW")
    for e in elements:
        links.append(e.get_attribute('href'))
    return(links)

web_driver = webdriver.Chrome()            # get a driver for Chrone
web_driver.set_window_size(1280, 720)      # we need a wide browser window to get the "text" link

# Do not change above this line
# --------------------------------------------------------


# get_links returns a list of URLs on the duolingos forum page.
# The forum page only loads a few posts to start with. So, we are giving 
# 10 second wait time so that YOU can scroll down the page to load more content.
print(get_links(web_driver, "https://forum.duolingo.com", 10))

# !!!!!
# !!!! Use this function to retrieve the URLs for posts, and save them into a file.
# If you save each URL per line in a file, you can read them back line by line. Then,
# use the codebe below (2nd code box) to retrieve the text and the streak.
# 
web_driver.close()                         # this will close the browser widnow.
````

````python
import urllib.request
from bs4 import BeautifulSoup as bs
import re
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

import time

# --------------------------------------------------------
# DO NOT CHANGE
#

# This function returns the text and the streak from a post page.
#
def get_content(driver, url):
    content = ""
    driver.get(url)
    time.sleep(4)

    icons = driver.find_elements_by_class_name("_21PEz")
    streak = int(icons[-1].text)
    
    post = driver.find_element_by_class_name("_2povu")
    content = post.text
    return streak, content

web_driver = webdriver.Chrome()            # get a driver for Chrone
web_driver.set_window_size(1280, 720)      # we need a wide browser window to get the "text" link

# Do not change above this line
# --------------------------------------------------------

# The following is a test. 
# You need to call this function for each URL.
#
streak, content = get_content(web_driver, "https://forum.duolingo.com/comment/15014194")

print("streak:  {}".format(streak))
print("content: {}".format(content))

web_driver.close()                         # this will close the browser widnow.
````
